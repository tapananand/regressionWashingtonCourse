```python
import pandas as pd
```

```python
info = pd.read_csv("../Philadelphia_Crime_Rate_noNA.csv")
```

```python
info
```

# Exploring the Data

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(info["CrimeRate"], info["HousePrice"], color="red")
plt.xlabel("CrimeRate")
plt.ylabel("HousePrice")
plt.show()
```

# Fit the regression model using CrimeRate as the feature

```python
X = info.iloc[:, 2:3].values
```

```python
y = info.iloc[:, 0:1].values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(X, y)
```

# Lets see what our fit looks like

```python
plt.scatter(X, y, color = "red")
plt.plot(X, regressor.predict(X), color = "blue")
plt.xlabel("CrimeRate")
plt.ylabel("HousePrice")
plt.show()
```

We see that our fitted line is leaning towards the kind of outlier observation on the extreme left (about 400 crimerate) - which has high crime rate but not so low house price as you would expect looking at the other data. So this observation is really influencing our model.

# Remove this point and redo the analysis

```python
info.sort_values("CrimeRate", ascending=False)
```

```python
info = info[info["MilesPhila"] != 0.0]
```

```python
info.sort_values("CrimeRate", ascending=False)
```

```python
X = info.iloc[:, 2:3].values
y = info.iloc[:, 0:1].values
```

```python
regressor2 = LinearRegression()
regressor2.fit(X, y)
```

```python
plt.scatter(X, y, color = "red")
plt.plot(X, regressor2.predict(X), color = "blue")
plt.xlabel("CrimeRate")
plt.ylabel("HousePrice")
plt.show()
```

Getting a much better fit now

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
regressor2.coef_
```

```python
regressor2.intercept_
```

We see that for the "no Center City" version, for per unit increase in crime, the predicted decrease in house prices is 2,288. In contrast, for the original dataset, the drop is only ~577 per unit increase in crime. This is significantly different!
